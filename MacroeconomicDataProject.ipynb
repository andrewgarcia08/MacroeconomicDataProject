- **Date**: The time period when the data was recorded (YYYY-MM-DD format).
- **M2_Money_Supply**: A measure of the total money supply that includes cash, checking deposits, and easily convertible near money (such as savings deposits and money market securities).
- **10Y Treasury Yield**: The return on investment for U.S. government debt securities (maturity = 10 years, often used as a benchmark for long-term interest rates.
- **Fed Funds Rate**: The interest rate at which depository institutions lend reserve balances to other banks overnight, set by the Federal Reserve.
- **CPI**: Consumer Price Index, which measures the average change over time in the prices paid by consumers for a market basket of goods and services.
- **Inflation_Rate_%**: The percentage change in the general price level of goods and services over a specific period, typically calculated from CPI data.
- **SOFR**: Secured Overnight Financing Rate, a benchmark interest rate that reflects the cost of borrowing cash overnight, collateralized by U.S. Treasury securities.

```python
plt.figure(figsize=(10,6))
plt.plot(df1['Date'], df1['Fed Funds Rate'], label='Fed Funds Rate')
plt.title('US Fed Funds Rate (Last 7 years)', fontsize=16) # graph title
plt.xlabel('Year') # set x-axis label to Year
plt.ylabel('Fed Funds Rate') # set y-axis label to Fed Funds Rate
plt.show()
```

This graph displays the US federal funds rate from the last 7 years.
From the data:

- __2018–2019__: Rates hovered between ~1.75% and 2.4%.

- __2020–2021__: Sharp drop to near 0% in early 2020, staying flat during the pandemic.

- __2022–2023__: Rapid rate hikes, peaking at about 5.25%.

- __2024–2025__: Slight decline toward ~4.75%.

```python
df2 = df1.copy()

df2['Yield_Spread'] = df2['10Y Treasury Yield'] - df2['Fed Funds Rate']
df2['Yield_Spread']
df2
```

```python
rolling_window = 30
for col in ['M2_Money_Supply', '10Y Treasury Yield', 'Fed Funds Rate', 'CPI', 'Inflation_Rate_%', 'SOFR']:
    df2[f'{col}_MA{rolling_window}'] = df2[col].rolling(window=rolling_window, min_periods=1).mean()
```

```python
df2
```

```python
corr = df2.corr(numeric_only=True)
corr.style.background_gradient()
```

```python
X = df2[['Fed Funds Rate']].copy() # independent variable 
Y = df2['SOFR'].copy() #dependent variable 

import statsmodels.api as sm
import math
X = sm.add_constant(X) 
est = sm.OLS(Y,X).fit()
print(est.summary())
print("\nAverage error: {:.2f}.".format(math.sqrt(est.mse_resid)))
```

```python
predictions = est.predict() # get predicted values

plt.ylabel('SOFR')
plt.xlabel('Fed Funds Rate')
plt.title('Fed Funds Rate vs. SOFR')
plt.plot(df2['Fed Funds Rate'], predictions, c='r')
plt.scatter(df2['Fed Funds Rate'],Y)
plt.show()
```
